### 1. Using the code from the lesson as a guide and the REST API from https://python.zach.lol/api/v1/items as we did in the lesson, create a dataframe named items that has all of the data for items.

```python
base_url = 'https://python.zach.lol'

#get response w/ requests.get()
response = requests.get(base_url)

#get api info 
response.json()
```

```python
#look into /documentation endpoint
response = requests.get(base_url + '/documentation')
response.json()
```

```python
#look at payload key
response = requests.get(base_url + "/documentation")
print(response.json()["payload"])
```

```python
#changing base url in order to explore items
base_url =  "https://python.zach.lol/api/v1"
items_response = requests.get(base_url + "/items")

items_response.json()
```

```python
#all info needed is in payload key
items_payload = items_response.json()['payload']
items_payload
```

```python
#items key has all the item info
items_payload.keys()
```

```python
#create a loop
items_list = []

response = requests.get('https://python.zach.lol/api/v1/items')
data = response.json()
n = data['payload']['max_page']

for i in range(1, n+1):
    url = 'https://python.zach.lol/api/v1/items?page=' + str(i)
    response = requests.get(url)
    data = response.json()
    page_items = data['payload']['items']
    items_list += page_items
```

```python
#create df that has all the data from items
items = pd.DataFrame(items_list)

items
```

### 3. Extract the data for sales (https://python.zach.lol/api/v1/sales). There are a lot of pages of data here, so your code will need to be a little more complex. Your code should continue fetching data from the next page until all of the data is extracted.

```python
#changing base url in order to explore stores
base_url =  "https://python.zach.lol/api/v1"
sales_response = requests.get(base_url + "/sales")

sales_response.json().keys()
```

```python
#all info needed is in payload key and sales key has all the sales info

sales_payload = sales_response.json()['payload']
sales_payload.keys()
```

```python
#create loop
base_url = 'https://python.zach.lol'

api_url = base_url + '/api/v1/'
response = requests.get(api_url + 'sales')
data = response.json()
    
# create list from 1st page
output = data['payload']['sales']

# loop through the pages and add to list
while data['payload']['next_page'] != None:
    
    response = requests.get(base_url + data['payload']['next_page'])
    data = response.json()
    output.extend(data['payload']['sales'])
    
sales = pd.DataFrame(output)
```

```python
sales
```

### 5. Combine the data from your three separate dataframes into one large dataframe.

```python
#merge sales w/ stores
combined_df = sales.merge(stores, left_on='store', right_on='store_id')

#merge sales and stores on items
combined_df = combined_df.merge(items, left_on='item', right_on='item_id')

#make complete dataframe to a csv
combined_df.to_csv('combined_df.csv')
```

```python
#alternate way
from io import StringIO

def pull_csv(url):
    '''
    This function pulls a .csv from a specified URL and returns a dataframe.
    '''
    req = requests.get(url)
    data = StringIO(req.text)
    df = pd.read_csv(data)
    df = pd.DataFrame(df)
    return df
```
